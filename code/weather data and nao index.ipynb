SOURCE: ROYAL NETHERLANDS METEOROLOGICAL INSTITUTE (KNMI)
Comment: These time series are inhomogeneous because of station relocations and changes in observation techniques. As a result, these series are not suitable for trend analysis. For climate change studies we refer to the homogenized series of daily data <http://www.knmi.nl/nederland-nu/klimatologie/daggegevens> or the Central Netherlands Temperature <http://www.knmi.nl/kennis-en-datacentrum/achtergrond/centraal-nederland-temperatuur-cnt>.

* YYYYMMDD  = Date (YYYY=year MM=month DD=day)
* DDVEC     = Vector mean wind direction in degrees (360=north, 90=east, 180=south, 270=west, 0=calm/variable)
* FHVEC     = Vector mean windspeed (in 0.1 m/s)
* FG        = Daily mean windspeed (in 0.1 m/s)
* FHX       = Maximum hourly mean windspeed (in 0.1 m/s)
* FHXH      = Hourly division in which FHX was measured
* FHN       = Minimum hourly mean windspeed (in 0.1 m/s)
* FHNH      = Hourly division in which FHN was measured
* FXX       = Maximum wind gust (in 0.1 m/s)
* FXXH      = Hourly division in which FXX was measured
* TG        = Daily mean temperature in (0.1 degrees Celsius)
* TN        = Minimum temperature (in 0.1 degrees Celsius)
* TNH       = Hourly division in which TN was measured
* TX        = Maximum temperature (in 0.1 degrees Celsius)
* TXH       = Hourly division in which TX was measured
* T10N      = Minimum temperature at 10 cm above surface (in 0.1 degrees Celsius)
* T10NH     = 6-hourly division in which T10N was measured; 6=0-6 UT, 12=6-12 UT, 18=12-18 UT, 24=18-24 UT
* SQ        = Sunshine duration (in 0.1 hour) calculated from global radiation (-1 for <0.05 hour)
* SP        = Percentage of maximum potential sunshine duration
* Q         = Global radiation (in J/cm2)
* DR        = Precipitation duration (in 0.1 hour)
* RH        = Daily precipitation amount (in 0.1 mm) (-1 for <0.05 mm)
* RHX       = Maximum hourly precipitation amount (in 0.1 mm) (-1 for <0.05 mm)
* RHXH      = Hourly division in which RHX was measured
* PG        = Daily mean sea level pressure (in 0.1 hPa) calculated from 24 hourly values
* PX        = Maximum hourly sea level pressure (in 0.1 hPa)
* PXH       = Hourly division in which PX was measured
* PN        = Minimum hourly sea level pressure (in 0.1 hPa)
* PNH       = Hourly division in which PN was measured
* VVN       = Minimum visibility; 0: <100 m, 1:100-200 m, 2:200-300 m,..., 49:4900-5000 m, 50:5-6 km, 56:6-7 km, 57:7-8 km,..., 79:29-30 km, 80:30-35 km, 81:35-40 km,..., 89: >70 km)
* VVNH      = Hourly division in which VVN was measured
* VVX       = Maximum visibility; 0: <100 m, 1:100-200 m, 2:200-300 m,..., 49:4900-5000 m, 50:5-6 km, 56:6-7 km, 57:7-8 km,..., 79:29-30 km, 80:30-35 km, 81:35-40 km,..., 89: >70 km)
* VVXH      = Hourly division in which VVX was measured
* NG        = Mean daily cloud cover (in octants, 9=sky invisible)
* UG        = Daily mean relative atmospheric humidity (in percents)
* UX        = Maximum relative atmospheric humidity (in percents)
* UXH       = Hourly division in which UX was measured
* UN        = Minimum relative atmospheric humidity (in percents)
* UNH       = Hourly division in which UN was measured
* EV24      = Potential evapotranspiration (Makkink) (in 0.1 mm)

```python
# load data
weather = pd.read_csv('Data/hourly weather data knmi.txt')

# strip spaces
weather.columns = weather.columns.str.strip()
weather = weather.applymap(lambda x: x.strip() if isinstance(x, str) else x)
weather.replace('', None, inplace=True)

# rename columns and select ones we want
column_mapping = {
    'YYYYMMDD':'Date',
    'UG':'DailyMeanHumidityPerc',
    'RH':'DailyPrecipitation',
    'RHX':'MaxHourlyPrecipitation',
    'RHXH':'MaxHourlyPrecipitationHour',
    'TG':'DailyMeanTemperature',
    'TN':'HourlyMinTemperature',
    'TNH': 'HourlyMinTemperatureHour',
    'TX': 'HourlyMaxTemperature',  
    'TXH': 'HourlyMaxTemperatureHour',  
    'FG' :'DailyMeanWindspeed',
    'FHX':'HourlyMaxMeanWindspeed',
    'FHXH':'HourlyMaxMeanWindspeedHour',
    'FHN' :'HourlyMinMeanWindspeed',
    'FHNH':'HourlyMinMeanWindspeedHour'
}

weather = weather[column_mapping.keys()]
weather.rename(columns=column_mapping, inplace=True)

# filter on only 2021 and later
weather = weather[weather['Date'] >= 20210101]
weather.reset_index(drop=True, inplace=True)

# cast date to datetime
weather['Date'] = pd.to_datetime(weather['Date'], format='%Y%m%d').dt.strftime('%Y/%m/%d')

hour_cols = [
    'MaxHourlyPrecipitationHour',
    'HourlyMinTemperatureHour',
    'HourlyMaxTemperatureHour',  
    'HourlyMaxMeanWindspeedHour',
    'HourlyMinMeanWindspeedHour'
]

for i in hour_cols:
    weather[i] = weather[i].astype(int)
    
weather['Date'] = pd.to_datetime(weather['Date'], format='%Y/%m/%d')
weather['Month'] = weather['Date'].dt.month
weather['Year'] = weather['Date'].dt.year
weather['YearMonth'] = weather['Date'].dt.to_period('M')
weather['FirstDayOfMonth'] = pd.to_datetime(weather[['Year', 'Month']].assign(DAY=1))
```

# Daily  and Monthly mean humidity percentage

```python
monthly_mean_humidity = weather.groupby('FirstDayOfMonth')['DailyMeanHumidityPerc'].mean().reset_index()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(weather['Date'], weather['DailyMeanHumidityPerc'], label='Daily Mean Humidity Percentage', alpha=0.7)
ax.plot(monthly_mean_humidity['FirstDayOfMonth'], monthly_mean_humidity['DailyMeanHumidityPerc'],
        marker='o', label='Monthly Mean Humidity Percentage')

ax.set_xlabel('Date')
ax.set_ylabel('Humidity Percentage')
ax.set_title('Daily and Monthly Mean Humidity Percentage')

plt.xticks(rotation=45, ha='right')  
plt.legend()
plt.grid(True)

plt.show()
```

### Daily in 2022

```python
plt.figure(figsize=(10, 6))
plt.plot(
    weather[weather['Date'].dt.year == 2022].Date, 
    weather[weather['Date'].dt.year == 2022].DailyMeanHumidityPerc, 
    linestyle='-', 
    color='b'
)

plt.title('Daily mean humidity percentage')
plt.xlabel('Date')
plt.ylabel('Humidity percentage')
plt.grid(True)
plt.show()
```

### Monthly in 2022

```python
monthly_humidity = weather[weather['Date'].dt.year == 2022].groupby('YearMonth')['DailyMeanHumidityPerc'].mean()
plt.figure(figsize=(10, 6))
plt.plot(monthly_humidity.index.astype(str), monthly_humidity.values, linestyle='-', color='b')

plt.title('Monthly mean humidity percentage in 2022')
plt.xlabel('Month')
plt.ylabel('Humidity percentage')
plt.grid(True)
plt.xticks(rotation=45, ha='right')  
plt.show()
```

# Hourly precipitation avg

```python
hourly_avg = weather.groupby('MaxHourlyPrecipitationHour')['MaxHourlyPrecipitation'].mean()

plt.figure(figsize=(10, 6))
plt.plot(hourly_avg.index, hourly_avg.values, linestyle='-', color='b')

plt.title('Hourly maximum precipitation')
plt.xlabel('Hour')
plt.ylabel('Precipitation')
plt.grid(True)
plt.show()
```

# Hourly maximum precipitation for each month

```python
monthly_avg = weather.groupby(['Month', 'MaxHourlyPrecipitationHour'])['MaxHourlyPrecipitation'].mean().reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Iterate over each month and plot the hourly average
for month, group in monthly_avg.groupby('Month'):
    ax.plot(group['MaxHourlyPrecipitationHour'], group['MaxHourlyPrecipitation'], label=f'Month {month}')

# Add labels and legend
ax.set_xlabel('Hour of Day')
ax.set_ylabel('Hourly Average Max Hourly Precipitation')
ax.set_title('Hourly Average Max Hourly Precipitation for Each Month')
ax.legend(title='Month')
plt.grid(True)

plt.show()
```

# NAO index

```python
NAO_file_path = 'Data/norm.nao.monthly.b5001.current.ascii.table.txt'

nao = pd.read_csv(NAO_file_path, delim_whitespace=True, index_col=0)
nao.reset_index(inplace=True)
nao_melted = pd.melt(nao, id_vars=['index'], var_name='Month', value_name='NAO')
nao_melted.rename(columns={'index':'Year'}, inplace=True)
month_mapping = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
             'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

nao_melted['Month'] = nao_melted['Month'].map(month_mapping)

# nao_melted['YearMonth'] = nao_melted['Year'].astype(str) + nao_melted['Month'].apply(lambda x: f'{x:02d}')
nao_melted['FirstDayOfMonth'] = pd.to_datetime(nao_melted[['Year', 'Month']].assign(DAY=1))
```

```python
nao_melted.sort_values('FirstDayOfMonth', inplace=True)
```

```python
nao_melted
```

```python
plt.figure(figsize=(10, 6))
plt.plot(nao_melted.FirstDayOfMonth, nao_melted.NAO, linestyle='-', color='b')

plt.title('NAO index')
plt.xlabel('Date')
plt.ylabel('NAO')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(nao_melted[nao_melted['Year'] == 2022].FirstDayOfMonth, nao_melted[nao_melted['Year'] == 2022].NAO, linestyle='-', color='b')

plt.title('Monthly NAO index in 2022')
plt.xlabel('Date')
plt.ylabel('NAO')
plt.grid(True)
plt.show()
```

```python
nao_melted['first_day_of_year'] = pd.to_datetime(nao_melted['Year'].astype(str) + '-01-01')
nao_melted['first_day_of_year'] = pd.to_datetime(nao_melted['first_day_of_year'], format='%Y/%m/%d')
yearly_avg_nao = nao_melted.groupby('first_day_of_year')['NAO'].mean().reset_index()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(nao_melted.FirstDayOfMonth, nao_melted.NAO, label='Monthly NAO index', alpha=0.7)
ax.plot(yearly_avg_nao.first_day_of_year, yearly_avg_nao.values, marker='o', label='Yearly avg NAO index')

ax.set_xlabel('Year')
ax.set_ylabel('NAO index')
ax.set_title('Monthly and Yearly NAO index')

plt.xticks(rotation=45, ha='right')  
plt.legend()
plt.grid(True)

plt.show()
```
